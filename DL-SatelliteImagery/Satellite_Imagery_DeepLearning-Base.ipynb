```python
import torch
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import os
from glob import glob

class SemanticSegmentationDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        """
        Args:
            root_dir (string): Root directory of the dataset.
                               e.g., '/root/.cache/kagglehub/.../Semantic-segmentation-dataset'
            transform (callable, optional): Optional transform to be applied on a sample.
        """
        self.root_dir = root_dir
        self.transform = transform
        self.image_paths = []
        self.mask_paths = []

        # Find all tile directories
        tile_dirs = glob(os.path.join(root_dir, 'Tile *'))

        # Iterate through each tile directory to find image and mask paths
        for tile_dir in tile_dirs:
            image_sub_dir = os.path.join(tile_dir, 'images')
            mask_sub_dir = os.path.join(tile_dir, 'masks')

            # Get image paths for the current tile
            image_files = sorted(glob(os.path.join(image_sub_dir, '*.jpg')))

            # Get corresponding mask paths
            mask_files = [os.path.join(mask_sub_dir, os.path.basename(f).replace('.jpg', '.png')) for f in image_files]

            # Extend the main lists
            self.image_paths.extend(image_files)
            self.mask_paths.extend(mask_files)

        print(f"Found {len(self.image_paths)} images and {len(self.mask_paths)} masks.")

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        # Load image and mask
        image_path = self.image_paths[idx]
        mask_path = self.mask_paths[idx]

        image = Image.open(image_path).convert("RGB")
        mask = Image.open(mask_path).convert("L")  # L for grayscale masks

        if self.transform:
            image, mask = self.transform(image, mask)

        return image, mask
```

```python
import numpy as np
import os
from glob import glob
from patchify import patchify
from sklearn.preprocessing import MinMaxScaler
from PIL import Image

# Define global variables for the dataset paths
dataset_root_folder = '/root/.cache/kagglehub/datasets/humansintheloop/semantic-segmentation-of-aerial-imagery/versions/1/Semantic-segmentation-dataset'
image_patch_size = 256
minmaxscaler = MinMaxScaler()

# Initialize the global lists to store the patched images and masks
image_dataset = []
mask_dataset = []

# Find all tile directoriestile_dirs = sorted(glob(os.path.join(dataset_root_folder, 'Tile *')))

# Iterate through each tile directory to find image and mask paths
for tile_dir in tile_dirs:
    image_sub_dir = os.path.join(tile_dir, 'images')
    mask_sub_dir = os.path.join(tile_dir, 'masks')

    image_files = sorted(os.listdir(image_sub_dir))

    for img_name in image_files:
        img_path = os.path.join(image_sub_dir, img_name)
        mask_path = os.path.join(mask_sub_dir, img_name.replace('.jpg', '.png'))

        if os.path.exists(img_path) and os.path.exists(mask_path):
            # Load image and mask using PIL
            image = Image.open(img_path).convert("RGB")

            # --- FIX IS HERE ---
            # Load the mask as an RGB image, not grayscale ("L")
            mask = Image.open(mask_path).convert("RGB")

            # Convert to numpy arrays for processing
            image = np.array(image)
            mask = np.array(mask)

            # Crop the images to be a multiple of the patch size
            size_x = (image.shape[1] // image_patch_size) * image_patch_size
            size_y = (image.shape[0] // image_patch_size) * image_patch_size
            image = image[0:size_y, 0:size_x]
            mask = mask[0:size_y, 0:size_x]

            # Patchify the images
            image_patches = patchify(image, (image_patch_size, image_patch_size, 3), step=image_patch_size)
            mask_patches = patchify(mask, (image_patch_size, image_patch_size, 3), step=image_patch_size)

            # Iterate through the patches and add them to the global lists
            for i in range(image_patches.shape[0]):
                for j in range(image_patches.shape[1]):
                    # Image processing
                    individual_patched_image = image_patches[i, j, 0, :, :, :]
                    individual_patched_image_scaled = minmaxscaler.fit_transform(
                        individual_patched_image.reshape(-1, individual_patched_image.shape[-1])
                    ).reshape(individual_patched_image.shape)
                    image_dataset.append(individual_patched_image_scaled)

                    # Mask processing
                    individual_patched_mask = mask_patches[i, j, :, :, :][0]
                    mask_dataset.append(individual_patched_mask)

# Convert the lists to numpy arrays for consistency and further use
image_dataset = np.array(image_dataset)
mask_dataset = np.array(mask_dataset)
print(f"Total number of patched images: {len(image_dataset)}")
print(f"Total number of patched masks: {len(mask_dataset)}")
```

```python
from keras.models import Model
from keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, Conv2DTranspose
from keras.layers import concatenate, BatchNormalization, Dropout, Lambda
```

```python
from keras import backend as K
```

```python
def jaccard_coef(y_true, y_pred):
  y_true_flatten = K.flatten(y_true)
  y_pred_flatten = K.flatten(y_pred)
  intersection = K.sum(y_true_flatten * y_pred_flatten)
  final_coef_value = (intersection + 1.0) / (K.sum(y_true_flatten) + K.sum(y_pred_flatten) - intersection + 1.0)
  return final_coef_value
```

```python
def multi_unet_model(n_classes=5, image_height=256, image_width=256, image_channels=1):

  inputs = Input((image_height, image_width, image_channels))

  source_input = inputs

  c1 = Conv2D(16, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(source_input)
  c1 = Dropout(0.2)(c1)
  c1 = Conv2D(16, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c1)
  p1 = MaxPooling2D((2,2))(c1)

  c2 = Conv2D(32, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(p1)
  c2 = Dropout(0.2)(c2)
  c2 = Conv2D(32, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c2)
  p2 = MaxPooling2D((2,2))(c2)

  c3 = Conv2D(64, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(p2)
  c3 = Dropout(0.2)(c3)
  c3 = Conv2D(64, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c3)
  p3 = MaxPooling2D((2,2))(c3)

  c4 = Conv2D(128, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(p3)
  c4 = Dropout(0.2)(c4)
  c4 = Conv2D(128, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c4)
  p4 = MaxPooling2D((2,2))(c4)

  c5 = Conv2D(256, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(p4)
  c5 = Dropout(0.2)(c5)
  c5 = Conv2D(256, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c5)

  u6 = Conv2DTranspose(128, (2,2), strides=(2,2), padding="same")(c5)
  u6 = concatenate([u6, c4])
  c6 = Conv2D(128, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(u6)
  c6 = Dropout(0.2)(c6)
  c6 = Conv2D(128, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c6)

  u7 = Conv2DTranspose(64, (2,2), strides=(2,2), padding="same")(c6)
  u7 = concatenate([u7, c3])
  c7 = Conv2D(64, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(u7)
  c7 = Dropout(0.2)(c7)
  c7 = Conv2D(64, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c7)

  u8 = Conv2DTranspose(32, (2,2), strides=(2,2), padding="same")(c7)
  u8 = concatenate([u8, c2])
  c8 = Conv2D(32, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(u8)
  c8 = Dropout(0.2)(c8)
  c8 = Conv2D(32, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c8)

  u9 = Conv2DTranspose(16, (2,2), strides=(2,2), padding="same")(c8)
  u9 = concatenate([u9, c1], axis=3)
  c9 = Conv2D(16, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(u9)
  c9 = Dropout(0.2)(c9)
  c9 = Conv2D(16, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c9)

  outputs = Conv2D(n_classes, (1,1), activation="softmax")(c9)

  model = Model(inputs=[inputs], outputs=[outputs])
  return model
```

## Generating Loss Function

- dice loss > Focal Loss > Total Loss
- Total Loss = (Dice loss + (1*Focal Loss))

```python
weights = [0.1666, 0.1666, 0.1666, 0.1666, 0.1666, 0.1666]
```

```python
import segmentation_models as sm
```

```python
dice_loss = sm.losses.DiceLoss(class_weights = weights)
```

```python
focal_loss = sm.losses.CategoricalFocalLoss()
```

```python
total_loss = dice_loss + (1 * focal_loss)
```
